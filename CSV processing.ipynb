```python
def getmax(array):
    (x,y) = array.shape
    maxval = 0
    for i in range(0,x):
        element = int(array[i][0])
        if (element>=int(maxval)):
            maxval = array[i][0]
    return int(maxval)
```

```python
class npElement:
    def __init__(self,nparray,index):
        self.index        = index
        self.phase        = int(nparray[0])
        self.task         = int(nparray[1])
        self.fe           = nparray[2]
        self.best         = float(nparray[3])
        self.lowest       = float(nparray[4])
        self.highest      = float(nparray[5])
        self.worst        = float(nparray[6])
        self.overrun      = float(nparray[7])
        self.rates        = int(nparray[8])
        self.source       = nparray[9]
        self.best_cost    = float(nparray[10])
        self.lowest_cost  = float(nparray[11])
        self.highest_cost = float(nparray[12])
        self.worst_cost   = float(nparray[13])
    
    def costToArray(self):
        return [self.best,self.lowest,self.highest,self.worst,self.best_cost,self.lowest_cost,self.highest_cost,self.worst_cost,self.rates]
    
    def printme(self):
        print("Debug---> ID:",self.index, " Phase: ",self.phase, " Task: ",self.task)
        
    def printPerson(self):
        print("ID: ",self.index, " Phase: ",self.phase, " ,Task: ",self.task, " ,Fee Earner: ", self.fe, "\n")
        
    def calculate_FF(self,debug_costs = False):
        self.lff = np.around((self.worst_cost-(self.highest_cost+self.lowest_cost)*0.5)*self.overrun+(self.highest_cost+self.lowest_cost)*0.5,-1)
        
        self.hff = np.around((self.worst_cost - self.highest_cost) * self.overrun + self.highest_cost,-1)
        
        if debug_costs==True:
            print(f'Highest Fixed Fee: {self.hff}')
            print(f'Lowest Fixed Fee: {self.lff}')
        
    def round_Costs(self,debug_costs = False):
        #debug_costs = True
        self.best_cost_R = int(np.around(self.best_cost,-1))
        self.lowest_cost_R = int(np.around(self.lowest_cost,-1))
        self.highest_cost_R = int(np.around(self.highest_cost,-1))
        self.worst_cost_R = int(np.around(self.worst_cost,-1))
        if debug_costs==True:
            print(f'Best cost:{self.best_cost} --> {self.best_cost_R}')
            print(f'Lowest cost:{self.lowest_cost} --> {self.lowest_cost_R}')
            print(f'Highest cost:{self.highest_cost} --> {self.highest_cost_R}')
            print(f'Worst cost:{self.worst_cost} --> {self.worst_cost_R}')
            print('---------------------------------------------------\n')
        
        
```

```python
#Precompile lists of npElements
#Format:
#         list[i][j]
#         i is used to denote the phase
#         j is used to denote the task in said phase

def compileList(nparray):
    listofPhases = []
    dummylist = []
    (x,y) = nparray.shape
    
    currentphase = 1

    for i in range(0,x):
        ele_temp = npElement(nparray[i],i)
        #print(nparray[i])
        phase = ele_temp.phase
        if (phase != currentphase):
            currentphase = phase
            listofPhases.append(dummylist)
            dummylist = []
        dummylist.append(ele_temp)
        
    listofPhases.append(dummylist)
        
    return listofPhases
```

```python
listOfIndicies = []

    
    

def compareCosts(ele_a,ele_b):
    namelist=["best","lowest","highest","worst","best_cost","lowest_cost","highest_cost","worst_cost"]
    match = False
    cost = []
    cost.append(ele_a.costToArray())
    cost.append(ele_b.costToArray())
#     print("cost[0]:", cost[0])
#     print("cost[1]:", cost[1])
#     print("Task: ",ele_a.task)
#     print("Match:",(cost[0]==cost[1]),"\n")
    if (cost[0] == cost[1]):
        return True
    else:
        print("ID: ", ele_a.index)
        print(f'Phase: {ele_a.phase}')
        print("Task: ",ele_a.task)
        print("Match:",(cost[0]==cost[1]),"\n")
        
        global listOfIndicies
        
        listOfIndicies.append(ele_a.index)
        
        
        for i in range(0,len(cost[0])-1):

            
            if (cost[0][i]!=cost[1][i]):
                name = namelist[i]
                print("Mismatch is found at: ",name)
                print("Values --->",ele_a.source,":",cost[0][i],"|",ele_b.source,":",cost[1][i],"\n")

        return False
```

```python
def compareTasks(biggerArray, smallerArray):
    allclear = True
    mismatches = []
    for element_b in biggerArray:
        found = False
        costsMatch = False
        
        
        for element_s in smallerArray:
            if (element_b.phase == element_s.phase):
                if (element_b.task == element_s.task):
                    if (element_b.fe == element_s.fe):
                        if (found == True):print("Dup found")
                        found = True
                        compareCosts(element_b,element_s)
                        
        if (found == False):
            mismatches.append(element_b)
    if (len(mismatches) > 0):
        global listOfIndicies
        
        print("\nPrinting mismatches in task: ")
        allclear = False
        for element in mismatches:
            listOfIndicies.append(element.index)
            element.printPerson()
    return allclear
```

```python
def algo_compute(listof_class):
    df_phase_costs = pd.DataFrame(columns=['phase','task','RBCC','RLAC','RHAC','RWCC','RLFF','RHFF','overrun'])
    current_index = 0
#     bestsum = 0
    for task_list in listof_class:
        for element in task_list:
            
            element.round_Costs()
            element.calculate_FF()
            df_phase_costs.loc[current_index] = [element.phase,element.task,element.best_cost_R,element.lowest_cost_R,element.highest_cost_R,element.worst_cost_R,element.lff,element.hff,element.overrun]
#             bestsum += element.best_cost_R
            current_index += 1
    return df_phase_costs
```
